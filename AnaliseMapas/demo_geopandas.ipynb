# Modelo Espacial de Dados

* Ponto: Usado por localização Pontual. Precisa de coordenadas (x,y,z=opcional)
* Linha: Usado pra trajetória. Composto por mais de um ponto e uma linha que conecta os pontos. É construido com um vetor de tuplas de coordenadas
* Polígono: Usado para Áreas, lagos e acidentes geográficos, calculo de proporções de áreas. Diversos pontos interligados em que o ultimo se liga com o primeiro

```python
from shapely.geometry import Point, LineString, Polygon
```

```python
ponto1 = Point(2,3)
ponto2 = Point(5,7)
ponto3 = Point(2,10)
```

```python
list(ponto1.coords)
```

```python
distancia = ponto1.distance(ponto2)
distancia
```

```python
linha = LineString([ponto1,ponto2,ponto3])
print(linha)
```

```python
#comprimento da linha
linha.length
```

```python
# ponto central (centroide)
linha.centroid
```

```python
# Para criar o poligos precisa passar diretamente as coordenadas, 
# nao pode ser um conjunto de linhas e pontos

poly = Polygon([(2,3),(5,7),(2,10)])
poly
```

```python
#obter as coordenadas do poligono
list(poly.exterior.coords)
```

```python
#calcular area
poly.area
```

```python
# pegar centroide do poligono
print(poly.centroid)
```

```python
# Extremidades do poligono
poly.bounds
```

```python
#coordenadas da linha exterior
poly.exterior
```

```python
#obter o tamanho da linha exterior
poly.exterior.length
```

# Operações entre Geometrias

```python
# criar dois circulos
circ1 = Point(1,1).buffer(1) #crescer o circulo em todas as extremidades
circ2 = Point(2,1).buffer(1) #Transforma o ponto em um poligono

circ1
```

```python
import matplotlib.pyplot as plt

x,y = circ1.exterior.xy
plt.plot(x,y)
x2,y2 = circ2.exterior.xy
plt.plot(x2,y2)
```

```python
# intersecao
inter = circ1.intersection(circ2)
inter
```

```python
#uniao
uniao = circ1.union(circ2)
uniao
```

```python
#diferença
dif = circ1.difference(circ2)
dif
```

```python
# operação de diferença simetrica
sim_dif = circ1.symmetric_difference(circ2)
sim_dif
```
